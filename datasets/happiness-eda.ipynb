```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=data, x="GDP_per_Capita", y="Happiness_Score", hue="Country")
plt.title("Happiness vs. GDP per Capita")
plt.show()
```

```python
plt.hist(data['Happiness_Score'], bins = 5)
plt.title('Ratings distribution')
plt.show()
```

```python
sns.lineplot(data=data, x='Happiness_Score', y='GDP_per_Capita')
```

```python
sns.lineplot(data=data, x='Social_Support', y='Happiness_Score')
```

```python
sns.lineplot(data=data, x='Happiness_Score', y='Unemployment_Rate')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Happiness_Score'], kde=True)
plt.title('Distribution of Happiness Scores', fontsize=16)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python

avg_happiness = data.groupby('Year')['Happiness_Score'].mean().reset_index()
plt.figure(figsize=(12,12))
plt.plot(avg_happiness['Year'], avg_happiness['Happiness_Score'], marker='o')
plt.title(' Average Happiness Score by Year')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.xticks(avg_happiness['Year'].unique())
plt.show()
```

```python
latest_year = data['Year'].max()
top_happiness_countries = data[data['Year'] == latest_year].sort_values('Happiness_Score', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness_Score', y='Country', data=top_happiness_countries, palette='viridis')
plt.title('Top 10 Happiest Countries in Latest Year', fontsize=16)
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```
